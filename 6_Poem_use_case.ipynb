## Scaling up

```python
# Copy path to a folder
folder = '/content/African-American-Literature-Text-Corpus-1853-1923'
```

```python
# gives us a nice progress bar
from tqdm.auto import tqdm # wrap tqdm() around any list or loop to see how fast you're looping
```

```python
# Loop over the files! Here's one way to do it

# 1. Start an empty list of (dictionaries)
counts_ld = []

# 2. Loop over files in your folder
for filename in tqdm(os.listdir(folder)):

    # (only consider text files)
    if filename.endswith('.txt'):

        # get the full filepath (folder + filename)
        filepath = os.path.join(folder,filename)

        # "step 1" read the file to a string
        fulltxt = open(filepath,errors='ignore').read()
        
        # "step 2" tokenize
        words = tokenize(fulltxt)
        
        # "step 3" count
        counts = Counter(words)
        
        # now: add a special key to be the index (leftmost column) of the dataframe
        counts['_filename'] = filename

        # add this dictionary to list of dictionaries of counts
        counts_ld.append(counts)

# 3. Finally, convert this list of dictionaries to a dataframe,
# where each dictionary becomes a row in a spreadsheet
counts_df = pd.DataFrame(counts_ld)
counts_df
```

```python
# let's call df the counts_df with _filename as its index
df = counts_df.set_index('_filename')
df
```

```python
# sums by column
df.sum()
```

```python
# flips dataframe
df.T
```

## Generating poetry?

```python
!pip install -qU transformers
from transformers import pipeline
import warnings
warnings.filterwarnings('ignore')

generator = pipeline('text-generation', model = 'gpt2')
generator
```

```python
generator("Because I could not stop for Death –", max_length = 30, num_return_sequences=3)
```

```python
generator("Hello, I'm a language model", max_length = 30, num_return_sequences=3)
```

```python
def gpt2(phrase):
    res = generator(phrase, num_return_sequences=1)
    return res[0].get('generated_text','')
```

```python
print(gpt2('''Because I could not stop for Death –
He kindly stopped for me –
'''))
```

```python
print(gpt2('''Because I could not stop for Death –
He kindly stopped for me –
'''))
```
